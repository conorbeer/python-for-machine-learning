## Visualize the data 

### Visualize the rating column with a histogram

- Note that almost all reviews fall between 2 and 4
- Most are rated 3-4
- It appears that chocolate is graded on a 1-5 scale

```python
import matplotlib.pyplot as plt


plt.figure(figsize = (16,5))
plt.hist(x = 'rating', data = dat_chocolate, bins = 4, color = "green")
plt.xlabel("Rating")
plt.show()
```

### Convert and visualize

- Change percent from string to numeric
- Plot the cocoa pct against rating


**Notes**

- It doesn't look like there's an obvious relationship between cocoa pct and rating

```python
dat_chocolate['cocoa_pct'] = dat_chocolate['cocoa_pct'].str.replace("%", "") # remove % sign

dat_chocolate['cocoa_pct'] = dat_chocolate['cocoa_pct'].astype(float) / 100 # convert to float and divide by 100 to yield fraction

head(dat_chocolate, 5)
```

```python
dat_chocolate.describe()

dat_chocolate.drop(dat_chocolate[dat_chocolate.cocoa_pct > 1].index, inplace=True)
```

```python
plt.figure(figsize = (16, 5))
plt.scatter('cocoa_pct', 'rating', data=dat_chocolate, color = 'black', alpha = 0.3)
plt.xlabel('cocoa_pct')
plt.ylabel('rating')
plt.show()


```
